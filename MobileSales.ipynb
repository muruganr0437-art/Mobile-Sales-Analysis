## Missing Value Treatment
Core and SSD specifications are missing mainly for mobile phones.
These values are filled with 'Not Applicable'.

```python
df['Core Specification'] = df['Core Specification'].fillna('Not Applicable')
df['SSD'] = df['SSD'].fillna('Not Applicable')
```

```python
df = df.drop_duplicates()
```

```python
df['Inward Date'] = pd.to_datetime(df['Inward Date'])
df['Dispatch Date'] = pd.to_datetime(df['Dispatch Date'])
```

```python
df['Price'] = df['Price'].astype(float)
```

```python
df.sample()
```

```python
df.nlargest(3,'Quantity Sold')
```

```python
df.nsmallest(3,'Quantity Sold')
```

## Feature Engineering
New features are created to improve business analysis.

```python
df['Revenue'] = df['Price'] * df['Quantity Sold']
df['Delivery Days'] = (df['Dispatch Date'] - df['Inward Date']).dt.days
```

## Exploratory Data Analysis (EDA)
This section explores sales, revenue, regional performance, and product characteristics
using visualizations.

## Revenue by Product Type
This graph shows which product category generates more revenue.

```python
product_revenue = df.groupby('Product')['Revenue'].sum()
product_revenue = product_revenue.sort_values(ascending=False)
plt.figure(figsize=(8,5))
bars = plt.bar(product_revenue.index, product_revenue.values, color='skyblue')
plt.title('Revenue by Product Type', fontsize=16)
plt.xlabel('Product Type', fontsize=12)
plt.ylabel('Total Revenue', fontsize=12)
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 0.02*height, f'{int(height)}', ha='center', fontsize=10)

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Top 10 Brands by Quantity Sold
This visualization highlights brands with the highest sales volume.

```python
top_qty_brands = (
    df.groupby('Brand')['Quantity Sold']
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

plt.figure(figsize=(10,5))
top_qty_brands.plot(kind='bar')
plt.title('Top 10 Brands by Quantity Sold')
plt.xlabel('Brand')
plt.ylabel('Quantity Sold')
plt.show()
```

## Revenue Distribution
This histogram shows how revenue values are distributed across transactions.

```python
sns.set(style="whitegrid")
fig, ax = plt.subplots(1, 1, figsize=(8, 4))
sns.histplot(
    df['Revenue'],
    bins=30,
    kde=True,     
    ax=ax
)
ax.set_title('Revenue Distribution')
ax.set_xlabel('Revenue')
ax.set_ylabel('Frequency')
plt.tight_layout()
plt.show()
```

## Region-wise Quantity Sold
This chart compares total quantity sold across different regions.

```python
region_qty = df.groupby('Region')['Quantity Sold'].sum()
fig, ax = plt.subplots(1, 1, figsize=(6, 6))
ax.pie(
    region_qty,
    labels=region_qty.index,
    autopct='%1.1f%%',
    startangle=90
)
ax.set_title('Quantity Sold Distribution by Region')
plt.tight_layout()
plt.show()
```

## Price vs RAM Size
This box plot shows how price varies with RAM size.

```python
df['RAM_GB'] = df['RAM'].str.replace('GB', '', regex=False).astype(int)
ram_values = sorted(df['RAM_GB'].unique())
fig, axes = plt.subplots(3, 3, figsize=(15, 12), sharey=True)
axes = axes.flatten() 
for ax, ram in zip(axes, ram_values):
    sns.boxplot(
        y=df[df['RAM_GB'] == ram]['Price'],
        ax=ax
    )
    ax.set_title(f'{ram} GB RAM')
    ax.set_xlabel('')
    ax.set_ylabel('Price')
for ax in axes[len(ram_values):]:
    ax.remove()
plt.suptitle('Price Distribution by RAM Size', fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

## Monthly Revenue Trend
This line chart shows revenue trends over time.

```python
df['Month'] = df['Dispatch Date'].dt.to_period('M')

monthly_sales = df.groupby('Month')['Revenue'].sum()

plt.figure(figsize=(12,5))
monthly_sales.plot()
plt.title('Monthly Revenue Trend')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.show()
```

## Correlation Heatmap
This heatmap shows relationships between numerical variables.

```python
plt.figure(figsize=(6,4))
sns.heatmap(
    df[['Price', 'Quantity Sold', 'Revenue', 'Delivery Days']].corr(),
    annot=True,
    cmap='coolwarm'
)
plt.title('Correlation Heatmap')
plt.show()
```

## Key Insights
- Laptops generate higher revenue compared to mobile phones
- Premium-priced products sell less but generate more revenue
- Central region contributes significantly to total sales
- Delivery time varies across regions, impacting customer experience
